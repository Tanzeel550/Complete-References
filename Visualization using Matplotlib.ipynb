### plot color, label

```python
plt.plot(x, np.sin(x), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')
# For short, you can use the following codes:plt.plot(x, x + 4, linestyle='-')  
# solidplt.plot(x, x + 5, linestyle='--') 
# dashedplt.plot(x, x + 6, linestyle='-.') 
# dashdotplt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
# we can use color and style at the same time
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

# 5) Scatter plots

### Scatter plot with plt.plot

```python
x = np.linspace(0,10,30)
y = np.sin(x)
```

```python
plt.plot(x,np.sin(x),'o', color='black')
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,label="marker='{0}'".format(marker))
    
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8)
```

```python
plt.plot(x, y, '-p', color='gray',markersize=15, linewidth=4,markerfacecolor='white',markeredgecolor='gray',markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```

### using plt.scatter

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,y,c= 'black')
# c is the color of markers
```

```python
plt.scatter(x,y, marker='*',s=100,cmap='virdis')
# if cmap is used then c must be omitted
# s is the size of marker
# different markers can be used like in above examples
```

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.randn(100)
size = 1000*np.random.randn(100)
plt.scatter(x,y, c = colors, s=size, cmap='rainbow',alpha=0.3,edgecolors='black')
# alpha is used to set the transperency of each marker
plt.colorbar()#color bar shows the color bar
```

# 6) Histograms, binnings & Density

```python
plt.hist(x)
```

```python
plt.hist(x, bins=50, alpha=0.5,histtype='bar', color='red',edgecolor='black')
```

```python
plt.hist(x, bins=30, alpha=0.5,histtype='bar', color='red',edgecolor='black',align='right',orientation="horizontal")
# align will allign the bins in the center, not commonly used
# orietation is used to set the orientation of the plot
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs)
```

### 2d historgrams

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]
x, y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues')
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
plt.hist2d(x, y, bins=30, cmap='Blues',cmin =10, cmax =20)
# cmin will not show bins below the cmin
# cmax will not show bins above the cmax
cb = plt.colorbar()
cb.set_label('counts in bin')
```

# 7) Customizing plot legends

```python
x = np.linspace(0,10,1000)
sinx = np.sin(x)
cosx = np.cos(x)
tanx = np.tan(x)
```

```python
plt.plot(x,sinx,label='sin(x)')
plt.plot(x,cosx,label='cos(x)')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x),label='Sine')
ax.plot(x, np.cos(x),label='Cosine')
leg = ax.legend()
```

```python
ax.legend(loc='best', frameon=False)
fig
# loc is used to fix the location of legend, in our case it is best
# it can be ['upper left', 'upper right', 'lower left', 'lower right', 'upper center', 'lower center', 'center left', 'center right']
# frameon : the name shows everything. if frameon is False, then frame of legend will not be shown
```

```python
ax.legend(loc='lower left', ncol=2)
fig
# ncol : We  can  use  the  ncol  command  to  specify  the  number  of  columns  in  the  legend
```

```python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
#fancybox is used to make rounded legend box
# framealpha is used to set the transperency of the box
# shadow is used to show the shadow of the box
# borderpad is used to add padding to the text inside the box
```

```python
ax.legend(['sin','cos'])
fig
```

```python
plt.figure()

plt.subplot(3,1,1)#(no of rows, no of column, current fig no.)
plt.plot(x,np.sin(x))

plt.subplot(3,1,2)
plt.plot(x,np.cos(x))

plt.subplot(3,1,3)
plt.plot(x,np.tan(x))
```

# 9) Customizing ticks

```python
y = np.linspace(0,10,10)
siny = np.sin(y)
cosy = np.cos(y)
tany = np.tan(y)
```

```python
fig = plt.figure()
ax = plt.axes()
plt.plot(y,siny)

ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.linspace(-1,1,10))
```

```python
num=np.arange(0, 3.14*2, 0.05)
fig=plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
sinnum=np.sin(num)
ax.plot(num, sinnum)

ax.set_xlabel('angle')
ax.set_title('sine')
ax.set_xticklabels(['zero','two','four','six'])
```

```python
fig = plt.figure()
ax = plt.axes()

plt.plot(y,siny)
ax.set_xticks(np.arange(0,10))
ax.set_xticklabels(['one','two','three','four','five','six','seven','eight','nine','ten','elevn'])
#note that xticklabels are correspondence to xticks and the same will happen for yticks

ax.set_yticks(np.linspace(-1,1,5))
ax.set_yticklabels(['-one', '-zero.five','zero','zero.five','one'])
```
